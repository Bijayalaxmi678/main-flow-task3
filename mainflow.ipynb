```python
import pandas as pd
```

```python
import matplotlib. pyplot as plt
```

```python
data = pd. read_csv("tips.csv")
```

```python
plt . scatter ( data['day'], data['tip'])
plt.title ( "Scatter Plot" )
plt.xlabel('Day')
plt.ylabel ('Tip' )
plt.show( )
```

```python
plt . scatter ( data['day'], data['tip'],s=data['total_bill'])
plt.title ( "Scatter Plot" )
plt.xlabel('Day')
plt.ylabel ('Tip' )
plt.colorbar ()
plt.show( )
```

```python
plt.plot (data['tip'])
plt.plot (data['size'])
# Adding Title to the Plot
plt.title( " Scatter Plot")
# Setting the X and Y labels
plt.xlabel('Day' )
plt.ylabel('Tip' )
plt.show( )
```

```python
plt.bar (data['day'], data ['tip'])
plt.title("Bar Chart")
# Setting the X and Y labels
plt.xlabel ('Day' )
plt.ylabel( 'Tip')
# Adding the legends
plt. show()
```
